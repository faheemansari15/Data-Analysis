```python
import pandas as pd
import numpy as np
```

```python
dropout = pd.read_csv("dropout-ratio-2012-2015.csv")
print(dropout)
```

```python
dropout.columns #show columns
```

```python
dropout.isna() #/See/find null/missing values (Shows all values False(Not Null) but there are some True(Null))
```

```python
dropout.isna().any() #If any value is Null gives True
```

```python
data_wb = dropout[dropout["State_UT"]=="West Bengal"]
data_wb.head() #But only 3 values
data_wb #So no use of head()
```

```python
data_wb.info()
```

```python
dropout["Primary_Boys"].values #Missing values are NR
```

```python
dropout = dropout.replace("NR", 0)
dropout["Primary_Boys"].values #Now, there are no missing values in dropout datasets
```

```python
dropout["Primary_Boys"] = dropout["Primary_Boys"].apply(lambda x: float(x))
dropout["Primary_Boys"].values #Now, All values of Primary_Boys are float
```

```python
dropout["Primary_Girls"] = dropout["Primary_Girls"].values.astype(np.float32)
dropout["Primary_Girls"].values
```

```python
dropout = dropout.replace("Uppe_r_Primary", 0)
dropout["Upper Primary_Boys"] = dropout["Upper Primary_Boys"].apply(lambda x: float(x)) 
dropout["Upper Primary_Boys"].values
```

```python
dropout["Upper Primary_Girls"] = dropout["Upper Primary_Girls"].apply(lambda x: float(x))
dropout["Upper Primary_Girls"].values
```

```python
dropout["HrSecondary_Boys"] = dropout["HrSecondary_Boys"].apply(lambda x: float(x)) 
dropout["HrSecondary_Boys"].values
```

```python
dropout["HrSecondary_Girls"] = dropout["HrSecondary_Girls"].apply(lambda x: float(x))
dropout["HrSecondary_Girls"].values
```
